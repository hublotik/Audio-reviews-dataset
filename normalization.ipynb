```python
from bs4 import BeautifulSoup
import unicodedata
import contractions
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re

STOPWORDS = set(stopwords.words('english'))

def clean_text(text, for_embedding=False):
    text = text.lower()
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text(separator=" ")
    text = contractions.fix(text)
    text = re.sub(r"'s\b", "", text)
    text = re.sub(r'\([^)]*\)', '', text)
    text = re.sub("[^a-zA-Z]", " ", text)

    if not for_embedding:
        text = " ".join([word for word in word_tokenize(text) if word not in STOPWORDS])
    text = re.sub(r'\b\w{1,2}\b', '', text)
    
    lemmatizer = WordNetLemmatizer()
    text = " ".join([lemmatizer.lemmatize(word) for word in word_tokenize(text)])

    return text

# Let's say that reviews is your list of reviews
def clean_reviews(reviews_list):
    return [clean_text(review) for review in reviews_list]

cleaned_reviews = clean_reviews(cleared_revs)
```

```python
def filter_list(data, key):
    # sort list in reverse order by key
    data.sort(key=lambda x: x[key], reverse=True)

    seen_keys = set()
    filtered_data = []

    # iterate over the list from largest to smallest
    for item in data:
        if item[key] not in seen_keys:
            # if we haven't seen this key before, add it to the set and keep the item
            seen_keys.add(item[key])
            filtered_data.append(item)

    return filtered_data

# Usage:


filtered_good_list = filter_list(good_list, 'id')
filtered_bad_list = filter_list(bad_list, 'id')
print(len(filtered_good_list))
```

```python
def filter_lists(list1, list2, attr1, attr2):
    # Create a dictionary from list2 with attr1 as the key and attr2 as the value.
    dict2 = {d[attr1]: d[attr2] for d in list2}

    filtered_list1 = []

    for item1 in list1:
        # If attr1 is not in both list1 and list2, keep the item from list1.
        if item1[attr1] not in dict2.keys():
            filtered_list1.append(item1)
        else:
            # Compare the attr2 values and only keep the item from list1 if it's not smaller.
            if item1[attr2] >= dict2[item1[attr1]]:
                filtered_list1.append(item1)

    return filtered_list1


```

```python
def compare_and_filter(lst):
    result = []
    unique_first_items = set()
    max_second_item = {}

    for sublist in lst:
        for item in sublist:
            first_item, second_item = item

            if first_item not in unique_first_items:
                unique_first_items.add(first_item)
                max_second_item[first_item] = second_item
            elif second_item > max_second_item[first_item]:
                max_second_item[first_item] = second_item

    for first_item, second_item in max_second_item.items():
        result.append((first_item, second_item))

    return result
pos_sim_res_filtered = compare_and_filter(pos_sim_res)
neg_sim_res_filtered = compare_and_filter(neg_sim_res)
```

```python
def compare_and_filter(list1, list2):
    result1 = []
    result2 = []

    for tuple1 in list1:
        found_match = False
        for tuple2 in list2:
            if tuple1[0] == tuple2[0]:
                found_match = True
                if tuple1[1] > tuple2[1]:
                    result1.append(tuple1)
                else:
                    result2.append(tuple2)
                break

        if not found_match:
            result1.append(tuple1)

    for tuple2 in list2:
        found_match = False
        for tuple1 in list1:
            if tuple2[0] == tuple1[0]:
                found_match = True
                break

        if not found_match:
            result2.append(tuple2)

    return result1, result2
final_result = compare_and_filter(pos_sim_res_filtered, neg_sim_res_filtered)
```
